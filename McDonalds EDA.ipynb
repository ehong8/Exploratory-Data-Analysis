## Units sold for the week separated into breakfast vs non-breakfast items

```python
sales['itemdesc'].unique()
```

```python
# List of breakfast item descriptions
breakfast_items = [
    'Total All Day Breakfast Items', 'Total Breakfast Sandwich', 'Total McMuffins',
    'Egg McMuffin', 'Sausage McMuffin', 'Sausage Egg McMuffin', 'Total Biscuits',
    'Bac Egg Ch Biscuit', 'Sausage Biscuit', 'Sausage Egg Biscuit', 'Hashbrown',
    'Total McGriddles', 'Bac Egg Ch McGriddle', 'Sau Egg Ch McGriddle', 'Total Bagels'
]

# Filter for breakfast items
all_day_breakfast_sales = sales[sales['itemdesc'].isin(breakfast_items)]

# Filter for non-breakfast items (everything not in the breakfast_items list)
non_breakfast_sales = sales[~sales['itemdesc'].isin(breakfast_items)]
```

```python
all_day_breakfast_units_sold = all_day_breakfast_sales.groupby('wk_ending').agg(
    total_adb_units_sold=('urws', 'sum')
).reset_index()

non_breakfast_units_sold = non_breakfast_sales.groupby('wk_ending').agg(
    total_non_breakfast_units_sold=('urws', 'sum')
).reset_index()

# Merging both datasets
units_sold_comparison = all_day_breakfast_units_sold.merge(
    non_breakfast_units_sold, on='wk_ending', how='left'
)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(units_sold_comparison['wk_ending'], units_sold_comparison['total_adb_units_sold'], label='All Day Breakfast Units Sold', color='g')
plt.plot(units_sold_comparison['wk_ending'], units_sold_comparison['total_non_breakfast_units_sold'], label='Non-Breakfast Units Sold', color='b')

plt.title('Units Sold Per Week: All Day Breakfast vs. Non-Breakfast Items Over Time')
plt.xlabel('Week Ending')
plt.ylabel('Units Sold')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Units Sold Per Week: All Day Breakfast versus Non-Breakfast Items Over Time')
plt.show()
```

## Item description: indivividual breakfast item performance

```python
# To break down individual item performance, we will filter the data for specific items such as McMuffins, Biscuits, and others.
# First, let's define the specific items to include in the breakdown.

item_groups = ['Total McMuffins', 'Total Biscuits', 'Total McGriddles', 'Total Bagels']

# Filter the sales data for these items
item_performance_data = sales[sales['itemdesc'].isin(item_groups)]

# Summarize units sold for each item group by week
item_performance_summary = item_performance_data.groupby(['wk_ending', 'itemdesc']).agg(
    total_units_sold=('urws', 'sum')
).reset_index()
```

```python
plt.figure(figsize=(10, 6))
for item in item_groups:
    item_data = item_performance_summary[item_performance_summary['itemdesc'] == item]
    plt.plot(item_data['wk_ending'], item_data['total_units_sold'], label=item)

plt.title('Units Sold for McMuffins, Biscuits, McGriddles, and Bagels Over Time')
plt.xlabel('Week Ending')
plt.ylabel('Units Sold')
plt.legend(title='Item Description')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Units Sold for McMuffins, Biscuits, McGriddles, and Bagels Over Time')
plt.show()
```

```python

# Manual label map for renaming columns
manual_label_map = {
    'adus': 'Average Daily Units Sold',
    'agc': 'Average Transactions per Day for that Week',
    'upt': 'Unit Sales Rate per 1000 Transactions',
    'wavg_price': 'Weighted Average Price',
    'urws': 'Units Sold for the Week'
}

# Check for null values in the dataset by county
null_counts_by_county = sales.groupby('County').apply(lambda x: x.isnull().sum())

# Filter for only numerical columns
numerical_columns = sales.select_dtypes(include=['float64', 'int64']).columns

# Drop unwanted columns: 'latitude', 'longitude', 'ItemN', and 'REST_KEY'
columns_to_remove = ['latitude', 'longitude', 'ItemN', 'REST_KEY', 'totunits']
numerical_columns = numerical_columns.drop(columns_to_remove, errors='ignore')

# Count null values by county for numerical columns
null_counts_by_county_numerical = sales.groupby('County')[numerical_columns].apply(lambda x: x.isnull().sum())

# Rename columns using the manual label map
null_counts_by_county_final = null_counts_by_county_numerical.rename(columns=manual_label_map)

# Plotting the heatmap of null values by county with rotated x and y axis labels
plt.figure(figsize=(14, 10))  # Increased figure size for better readability
sns.heatmap(null_counts_by_county_final, cmap='YlGnBu', annot=True, fmt="g", cbar=False)

# Rotate x-axis and y-axis labels for better fit
plt.title('County-wise Distribution of Null Values in Key Sales Metrics', fontsize=18, weight = 'bold')
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate x-axis labels (variable names)
plt.yticks(rotation=45, ha='right', fontsize=10)  # Rotate y-axis labels (county names)
plt.ylabel('County', fontsize = 14, weight = 'bold')

plt.tight_layout()
plt.savefig('County-wise Distribution of Null Values in Key Sales Metrics')
plt.show()
```
